## Letra A)

Exibir na Tela (Precisão de uma (1) casa decimal: .1f): Média Individual, 1° Bimestre, 2° Bimestre, Semestral e Final (após a Prova Final).

Nota da Prova Final é lida (**INPUT**) do usuário, somente para os que estão de Prova Final.

```
# USAR OS CÓDIGOS ABAIXO PARA SIMULAR AS NOTAS DE ENTRADA DAS PROVAS FINAIS
# Média semestral (sem prova final) = (1° Bimestre + 2° Bimestre) / 2
```

```
# Preenchimento automático da Prova Final
media = (dataset4[:,1] + dataset4[:,2]) / 2
media[:5]
```

```
# Alunos de Prova Final
indexProvaFinal = ((media >= 3) & (media < 7))
sum(indexProvaFinal)
```

```
ProvaFinal = np.zeros(len(dataset4))
mean = 5.0
std = 0.9
ProvaFinal[indexProvaFinal] = np.random.normal(mean, std, sum(indexProvaFinal))
```

```
ProvaFinal
```

```
mediaFinal = (media + ProvaFinal) / 2
mediaFinal
```

```
print('PAUTA: DADOS DOS 5 PRIMEIROS ALUNOS:\n')
for indice, aluno in enumerate(dataset4[:5]):
  print(f'ALUNO {indice + 1}:')
  print(f'Matrícula          : {aluno[0]: .0f}')
  print(f'Nota do 1º Bimestre: {aluno[1]: .1f}')
  print(f'Nota do 2º Bimestre: {aluno[2]: .1f}')
  print(f'Media Semestral: {media[indice]: .1f}')
  if(media[indice] < 7):
    print(f'Nota Prova Final: {ProvaFinal[indice]: .1f}')
    print(f'Nota Media Final: {mediaFinal[indice]: .1f}')
  print('\n')
```
